```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

column_names = [
    "edible/poisonous",
    "cap-shape",
    "cap-surface",
    "cap-color",
    "bruises",
    "odor",
    "gill-attachment",
    "gill-spacing",
    "gill-size",
    "gill-color",
    "stalk-shape",
    "stalk-root",
    "stalk-surface-above-ring",
    "stalk-surface-below-ring",
    "stalk-color-above-ring",
    "stalk-color-below-ring",
    "veil-type",
    "veil-color",
    "ring-number",
    "ring-type",
    "spore-print-color",
    "population",
    "habitat"
]

df = pd.read_csv('./data/agaricus-lepiota.data', header=None, names=column_names)

df
```

```python
# Bar chart for the 'cap-shape' column
df['cap-shape'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Cap Shape')
plt.xlabel('Cap Shape')
plt.ylabel('Count')
plt.show()
```

```python
# Grouped bar chart for 'cap-shape' and 'cap-color'
pd.crosstab(df['cap-shape'], df['cap-color']).plot(kind='bar', stacked=True)
plt.title('Cap Shape vs Cap Color')
plt.xlabel('Cap Shape')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of edible vs inedible (poisonous)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='habitat', hue='edible/poisonous', palette='Set2')
plt.title('Distribution of Habitat between Edible and Poisonous Mushrooms')
plt.xlabel('Habitat')
plt.ylabel('Count')
plt.show()
```
